# Part II - In-Depth Analysis

In this section, we delve deeper into the dataset to uncover more nuanced insights. We'll explore temporal patterns by user type, examine how different age groups use the bike-sharing system, and conduct a multivariate exploration to understand the relationships between multiple variables. This comprehensive analysis aims to provide actionable insights for enhancing the bike-sharing service.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the clean dataset using raw string notation for the file path
df_clean = pd.read_csv(r'C:\Users\dariu\OneDrive\Desktop\explore\df_clean.csv')

# Convert start_time and end_time to datetime format
df_clean['start_time'] = pd.to_datetime(df_clean['start_time'])
df_clean['end_time'] = pd.to_datetime(df_clean['end_time'])

# Ensure categorical columns are of type 'category'
df_clean['user_type'] = df_clean['user_type'].astype('category')
df_clean['member_gender'] = df_clean['member_gender'].astype('category')

# Drop rows with any missing values
df_clean.dropna(inplace=True)
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df_clean, x=df_clean['start_time'].dt.hour, hue='user_type', palette='viridis')
plt.title('Hourly Usage by User Type')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Trips')
plt.xticks(range(0, 24))
plt.legend(title='User Type')
plt.show()
```

## Hourly Usage by User Type

This analysis separates the bike usage by hour for different user types, namely Subscribers and Customers. The distinct patterns between the two groups could indicate differing usage motivations, such as commuting for Subscribers and more leisure or erratic usage times for Customers. Understanding these differences can help tailor marketing and operational strategies to each user segment's needs.


```python
# Assuming age has been calculated and stored in 'age' column
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_clean, x='user_type', y='age', palette='viridis')
plt.title('Age Distribution by User Type')
plt.xlabel('User Type')
plt.ylabel('Age')
plt.show()
```

## Age Distribution by User Type

This box plot shows the age distribution among different user types. By comparing the age ranges and medians between Subscribers and Customers, we can infer demographic trends that might influence service preferences and usage patterns. Such insights are valuable for targeted marketing and enhancing service offerings to meet diverse user needs.


```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df_clean, x='age', y='duration_minute', hue='user_type', style='member_gender', alpha=0.5)
plt.title('Trip Duration by Age and User Type')
plt.xlabel('Age')
plt.ylabel('Duration (Minutes)')
plt.xlim(18, 80)  # Assuming most users are between 18 and 80 years old
plt.ylim(0, 60)  # Focusing on trips under 60 minutes for clarity
plt.legend(title='User Type & Gender')
plt.show()
```

## Multivariate Exploration: Trip Duration by Age and User Type

This scatter plot explores the relationship between trip duration, age, and user type, with different markers for gender. The visualization aims to uncover patterns that might exist among these variables, such as whether younger or older users tend to take longer trips or how trip durations vary between subscribers and customers. Analyzing these relationships can provide deeper insights into user behavior and preferences, guiding service improvements and user engagement strategies.
